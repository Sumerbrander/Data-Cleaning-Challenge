**Dealing with NaN and unnecessary columns**

**Replacing values**

```python
#Let's observe the value of the columns with monetary information.
fifa[['Release Clause', 'Value', 'Wage']]
```

We will replace M and K with the appropriate values for columns Wage, Value and Release Clause. We will also remove all currency and symbols.

```python
# Convert the column to a string first
fifa['Wage'] =fifa['Wage'].astype(str)

# Replace "M" with "000000" and "K" with "000"
fifa['Wage'] = fifa['Wage'].str.replace('M', '000000').str.replace('K', '000')

# Clean the column by removing any currency symbols and commas
fifa['Wage'] = fifa['Wage'].str.replace('€', '').str.replace(',', '').astype(float)
```

```python
# Convert the column to a string first
fifa['Release Clause'] =fifa['Release Clause'].astype(str)

# Replace "M" with "000000" and "K" with "000"
fifa['Release Clause'] = fifa['Release Clause'].str.replace('M', '000000').str.replace('K', '000')

# Clean the column by removing any currency symbols and commas
fifa['Release Clause'] = fifa['Release Clause'].str.replace('€', '').str.replace(',', '').str.replace('.', '').astype(float)
```

```python
fifa_value = fifa.copy()

fifa_value.loc[1, 'Value']

fifa_value.loc[1, 'Value'][1:-1]

for i in range(len(fifa_value['Value'])):
    if fifa_value.loc[i, 'Value'][1] == '0':
       fifa_value.loc[i, 'Value'] = 0
    else:
        fifa_value.loc[i, 'Value'] = fifa_value.loc[i, 'Value'][1:-1]
 
fifa_value['Value'] = fifa_value['Value'].astype('float')

fifa_value['Value'] = fifa_value['Value'] * 1000000

fifa_value.head()
```

As we can see above, all values now ends with zeros(0) and all symbols has been removed.


Next, We will check the Heights column if they are all in cm

```python
# Check if all values in the 'Height' column contain 'cm'
all_values_contain_cm = fifa_value['Height'].str.contains('cm').all()

if all_values_contain_cm:
    print("All values in the 'Height' column contain 'cm'")
else:
    print("Not all values in the 'Height' column contain 'cm'")
```

```python
print(fifa_value['Height'].unique())
```

Some values are in feet and inches not cm, so we write a function that converts everything to cm.

```python
def convert_height(height):
    if 'cm' in height:
        return float(height.replace('cm', ''))
    elif '\'' in height:
        feet, inches = height.split('\'')
        total_inches = int(feet) * 12 + int(inches.replace('"', ''))
        return round(total_inches * 2.54, 1)
    else:
        return None

# Apply the function to the 'Height' column
fifa_value['Height'] = fifa_value['Height'].apply(convert_height)
```

```python
fifa_value['Height'].head(25)
```

Let's also observe the weight column if they are all in Kg.

```python
# Check if all values in the 'Weight' column contain 'kg'
all_values_contain_kg = fifa_value['Weight'].str.contains('kg').all()

if all_values_contain_kg:
    print("All values in the 'Height' column contain 'kg'")
else:
    print("Not all values in the 'Height' column contain 'kg'")
```

```python
print(fifa_value['Weight'].unique())
```

Some values are in lbs,so we convert them to kg first and then remove the kg to have a strictly numeric column

```python
# Convert values in pounds to kilograms
mask = fifa_value['Weight'].str.contains('lbs')
fifa_value.loc[mask, 'Weight'] = fifa_value.loc[mask, 'Weight'].apply(lambda x: str(round(float(x.replace('lbs', '')) * 0.453592, 2)) + 'kg')
```

```python
# Create a new column to store weight in kilograms
fifa_value['Weight(kg)'] = fifa_value['Weight']

# Define a function to convert weight to kilograms
def to_kg(weight):
    if 'lbs' in weight:
        weight = float(weight.replace('lbs', '').strip()) * 0.453592
    elif 'kg' in weight:
        weight = float(weight.replace('kg', '').strip())
    return weight

# Apply the to_kg function to the Weight column
fifa_value['Weight(kg)'] = fifa_value['Weight(kg)'].apply(to_kg)

# Convert the Weight(kg) column to float datatype
fifa_value['Weight(kg)'] = fifa_value['Weight(kg)'].astype(float)

fifa_value['Weight'] = fifa_value['Weight'].str.replace('kg', '')

```

We can observe that the columns; W/F, SM and IR contains the star sign. We will go ahead and remove them.

```python
# Remove special characters that look like stars
fifa['W/F'] = fifa['W/F'].replace(to_replace=r'[^0-9]', value='', regex=True)
fifa['SM'] = fifa['SM'].replace(to_replace=r'[^0-9]', value='', regex=True)
fifa['IR'] = fifa['IR'].replace(to_replace=r'[^0-9]', value='', regex=True)
fifa[['W/F', 'SM', 'IR']].head(10)
```
